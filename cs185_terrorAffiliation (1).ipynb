# Analysis of Network
Degree of Centrality
Most Influential
Most important connection
Best betweener
Shortest Path



Simple method to display heat map of centrality measures
Yellow/Orange color of nodes represent high centrality measure
Blue/Purple color of nodes represent low centrality measure

```python



#This function takes a graph, position of nodes, the centrality
def draw (mygraph, pos, centralityMeasure, centralityTitle):
    
    nodes = nx.draw_networkx_nodes(mygraph, pos, node_size=50, cmap=plt.cm.plasma, 
                                   node_color=list(centralityMeasure.values()),
                                   nodelist=centralityMeasure.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1, base=25))
    edges = nx.draw_networkx_edges(mygraph, pos)

    plt.title(centralityTitle)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
    
```

```python
#General connection degree of all nodes
nx.degree(mygraph)
```

```python
pos = nx.spring_layout(mygraph, seed=675)
draw(mygraph, pos, nx.degree_centrality(mygraph), 'Degree Centrality')
```

```python
#Most influential, centrality of each node
nx.degree_centrality(mygraph)
```

```python
#Get the most influential node in order
most_influential = nx.degree_centrality(mygraph)
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

```python
#Most important connection, eigenvector
nx.eigenvector_centrality(mygraph)
```

```python
pos = nx.spring_layout(mygraph, seed=675)
draw(mygraph, pos, nx.eigenvector_centrality(mygraph), 'Eigenvector')

```

```python
#Find most important bridge in descending order, Mohamed Atta most important
most_important_bridge = nx.eigenvector_centrality(mygraph)
for node in sorted(most_important_bridge, key=most_important_bridge.get, reverse=True):
    print(node, most_important_bridge[node])
```

```python
#Closeness centrality, what is the closest connection path between node1 and node2

print("Shortest connection between Mohamed Atta & Osama Awdallah")
print(nx.shortest_path(mygraph,"Mohamed Atta","Osama Awadallah"))
print("Shortest connection between Mohamed Atta & Marvan-Al-Shehhi")
print(nx.shortest_path(mygraph,"Mohamed Atta","Marwan Al-Shehhi"))

```

```python
#Betweenness centrality, we want to find out who is the best at connecting nodes, 
#number of times a node acts as a bridge along the shortest path between two other nodes.

best_connecting_node = nx.betweenness_centrality(mygraph)
for w in sorted(best_connecting_node, key=best_connecting_node.get, reverse=True):
    print(w, best_connecting_node[w])
```

```python

pos = nx.spring_layout(mygraph, seed=675)
draw(mygraph, pos, nx.betweenness_centrality(mygraph), 'Betweenness')


```
